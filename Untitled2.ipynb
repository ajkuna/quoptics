```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from numpy import *
from qutip import *
from mpl_toolkits.mplot3d import Axes3D   
from matplotlib import cm   
from pylab import *  

N = 4
chi = 1*2*pi
t = linspace(0,10, 1000)

a = destroy(N)
n = num(N)

# the Kerr Hamiltonian
H = 0.5 * chi * a.dag() * a.dag() * a * a

# we start with a coherent state with alpha=2.0
xvec = linspace(-2,2,17)
yvec = linspace(-2,2,17)
psi0 = coherent(4, 0)
rho = coherent_dm(4,0)

Q_coherent_state = qfunc(psi0, xvec,yvec)

Q_coherent = qfunc(rho, xvec, xvec)

fig, axes = plt.subplots(1, 2, figsize=(12,3))
#fig1, axes1 = plt.subplots(1, 2, figsize=(12,3))
cont0 = axes[0].contourf(xvec, xvec, Q_coherent_state, 100)
cont1 = axes[1].contourf(xvec, xvec,Q_coherent,100)
lbl0 = axes[0].set_title("Coherent state")
lbl1 = axes[1].set_title("Coherent state dm")
#line0 = axes1[0].plot(xvec, Q_coherent_state)

for i in range(len(yvec)):
    if yvec[i] == 0:
        print(i)
        fig1, axes1 = plt.subplots(1, 2, figsize=(12,3))
        line0 = axes1[0].plot(xvec, Q_coherent_state[i])
        plt.show()

plt.show()
```

```python
# we start with a coherent state with alpha=2.0
psi1 = coherent(4, 2)
# and evolve the state under the influence of the hamiltonian. 
# by passing an empty list as expecation value operators argument, 
# we get the full state of the system in result.states
result = mesolve(H, psi1, t, [], [])
```
