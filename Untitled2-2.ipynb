```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from numpy import *
from qutip import *
from mpl_toolkits.mplot3d import Axes3D   
from matplotlib import cm   
from pylab import *  


# Define constants 
N = 20 # Hillbert space
chi = 1*2*pi  # Kerr constant
t = linspace(0,10,1000)

a = destroy(N) # Annihilation operator
n = num(N)     # Number operator

# the Kerr Hamiltonian
H = 0.5 * chi * a.dag() * a.dag() * a * a

# We start with a coherent state with alpha=0

psi0 = coherent(4, 0)
rho = coherent_dm(4,0)




# Define q function range
N = 20 # size of the Hilbert space 
xvec2 = linspace(-2,2,17) 

alpha = []
for n in linspace(-2,2,17):
    for m in linspace(-2,2,17):
        perm_a = i + m*1j
        alpha.append(perm_a)

def husimi(x):
    psi_new=(displace(20,x)*basis(20,0)).unit();
    dm_psi_new = ket2dm(psi_new)
    q_new= 1/pi*(inner(psi_new.dag(),dm_psi_new))
    return q_new

q = parfor(husimi, alpha)




xvec = linspace(-2,2,17)
yvec = linspace(-2,2,17)


Q_coherent_state = qfunc(psi0, xvec,yvec)

Q_coherent = qfunc(rho, xvec, xvec)

fig, axes = plt.subplots(1,2 , figsize=(4,4))
#axes.contourf(xvec, xvec, Q_coherent_state,100)
cont0 = axes[0].contourf(xvec, xvec, Q_coherent_state, 100)
cont1 = axes[1].contourf(xvec2, xvec2, asarray(q),100)
lbl0 = axes[0].set_title("Coherent state")
lbl1 = axes[1].set_title("Coherent state dm")


for i in range(len(yvec)):
    if yvec[i] == 0:
        print(i)
        fig1, axes1 = plt.subplots(1, 1, figsize=(4,4))
        line0 = axes1.plot(xvec, Q_coherent_state[i])
        plt.show()

plt.show()




```

```python
# we start with a coherent state with alpha=2.0
psi1 = coherent(4, 2)
# and evolve the state under the influence of the hamiltonian. 
# by passing an empty list as expecation value operators argument, 
# we get the full state of the system in result.states
result = mesolve(H, psi1, t, [], [])
```

```python
# New Husimi space

xvec1 = linspace(-5,5,200)

q_result = qfunc(result.states[1],xvec1,xvec1)
q_result2 = qfunc(result.states[3],xvec1,xvec1)


fig2, axes2 = plt.subplots(1, 2, figsize=(12,3))
#fig1, axes1 = plt.subplots(1, 2, figsize=(12,3))
cont0 = axes2[0].contourf(xvec1, xvec1, q_result, 100)
cont1 = axes2[1].contourf(xvec1, xvec1,q_result2,100)
lbl0 = axes2[0].set_title("Husimi Q, state 1")
lbl1 = axes2[1].set_title("Husimi Q, state 2")


plt.show()


```
