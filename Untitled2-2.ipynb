```python
t = linspace(0,10,1000)
# we start with a coherent state with alpha=2.0
kappa = .001
psi1 = coherent(20, 2)
# the Kerr Hamiltonian
H = 0.5 * chi * a.dag() * a.dag() * a * a 
c_op = []
c_op.append(sqrt(kappa) * a)
# and evolve the state under the influence of the hamiltonian. 
# by passing an empty list as expecation value operators argument, 
# we get the full state of the system in result.states
result = mesolve(H, psi1, t, c_op, [])
```

```python
# New Husimi space

# xvec1 = linspace(-5,5,200)

# q_result = qfunc(result.states[20],xvec1,xvec1)
q_result2 = zeros((len(alpha), len(alpha)))
for i in range(len(alpha)):
    for j in range(len(alpha)):
        first = husimi2(alpha[i,j], result.states[40], 20)
        second = first[0,]
        third = second[0,]
#         print (first[0,])
        q_result2[i,j] = third[0,]
#         print(q_result2)

fig2, axes2 = plt.subplots(1, 2, figsize=(12,3))
#fig1, axes1 = plt.subplots(1, 2, figsize=(12,3))
# cont0 = axes2[0].contourf(alpha, alpha, q_result, 100)
cont1 = axes2[1].contourf(alpha.real, alpha.imag, q_result2,100)
# lbl0 = axes2[0].set_title("Husimi Q, state 1")
lbl1 = axes2[1].set_title("Husimi Q, state 2")

plt.show()
```
